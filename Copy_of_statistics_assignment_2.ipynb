Question 1: What is a random variable in probability theory?

Answer 1: In probability theory, a random variable is a variable that takes on numerical values determined by the outcome of a random experiment. Essentially, it is a way to assign numbers to the results of a random process.

example : tossing a coin







Question 2: What are the types of random variables?

Answer 2 :

1.Discrete Random Variables

Definition: A discrete random variable takes countable values, usually integers.

Characteristics:

Can list all possible values.

Often counts things like number of successes, number of arrivals, etc.

Examples:

Number of heads in 5 coin tosses.

Number of students present in a class.

Number of cars passing through a toll booth in an hour.

2.Continuous Random Variables

Definition: A continuous random variable takes any value in a given interval (uncountable values).

Characteristics:

Values are measured, not counted.

Probability is calculated over an interval rather than at a single point.

Examples:

Height of students in a school.

Time required to complete a task.

Temperature in a city on a particular day.

Question 3: Explain the difference between discrete and continuous distributions.

Answer 3:
1. Discrete Distribution

Definition: A distribution that shows probabilities of a discrete random variable taking specific, countable values.

Characteristics:

Values are countable (e.g., 0, 1, 2, …).

Probability is assigned to each specific value.

Sum of all probabilities = 1.

Examples:

Probability of getting 0, 1, 2… heads in 3 coin tosses.

Number of students passing an exam.

Probability Representation: Often shown in a probability mass function (PMF).

2. Continuous Distribution

Definition: A distribution that shows probabilities of a continuous random variable taking values in an interval.

Characteristics:

Values are uncountable; any number in a range is possible.

Probability of a single exact value is 0; we calculate probability over an interval.

Total area under the curve = 1.

Examples:

Probability that a student’s height is between 150 cm and 160 cm.

Time taken to complete a task.

Probability Representation: Often shown using a probability density function (PDF).

| Feature                    | Discrete Distribution            | Continuous Distribution            |
| -------------------------- | -------------------------------- | ---------------------------------- |
| Values                     | Countable                        | Uncountable (any in a range)       |
| Probability of exact value | Non-zero                         | Zero                               |
| Representation             | PMF (Probability Mass Function)  | PDF (Probability Density Function) |
| Example                    | Number of heads in 3 coin tosses | Height of a person                 |

Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5.

Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.


```python
''' Answer 9 '''
import numpy as np
import matplotlib.pyplot as plt
mean = 50
std_dev = 5
n = 1000

# Generate random numbers from a normal distribution
random_numbers = np.random.normal(loc=mean, scale=std_dev, size=n)
calculated_mean = np.mean(random_numbers)
calculated_std_dev = np.std(random_numbers)
print(f"Mean: {calculated_mean:.2f}")
print(f"Standard Deviation: {calculated_std_dev:.2f}")
# Draw a histogram to visualize the distribution
plt.figsize = (10, 6)
plt.hist(random_numbers, bins=30, density=True, alpha=0.6, color='g', edgecolor = 'black')
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(linestyle='--', alpha=0.7)
plt.show()



```

Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.


```python
import numpy as np
from scipy import stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Sample size
n = len(daily_sales)

# Sample mean
mean_sales = np.mean(daily_sales)

# Sample standard deviation
std_sales = np.std(daily_sales, ddof=1)  # ddof=1 for sample standard deviation

# Standard error
SE = std_sales / np.sqrt(n)

# 95% confidence interval using t-distribution
confidence = 0.95
t_value = stats.t.ppf((1 + confidence) / 2, df=n-1)  # two-tailed

# Margin of error
ME = t_value * SE

# Confidence interval
CI_lower = mean_sales - ME
CI_upper = mean_sales + ME

# Print results
print(f"Mean Daily Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: [{CI_lower:.2f}, {CI_upper:.2f}]")
```
